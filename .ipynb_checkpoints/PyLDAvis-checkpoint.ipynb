```python
import json
import pyLDAvis
#import pyLDAvis.gensim  # don't skip this
import matplotlib.pyplot as plt
%matplotlib inline
import pyLDAvis.gensim_models 
import gensim
import gensim.corpora as corpora
```

```python
import pyLDAvis.gensim_models
import pickle 
import pyLDAvis
# Visualize the topics
year = 1850
with open(r'C:\Users\user\Desktop\UNIMI\ProjectInfoRet\data\doc_token_{}.json'.format(year)) as json_file:
    ls_opinioni_tokenizzate = json.load(json_file)['ls_opinioni_tokenizzate']
    
id2word = corpora.Dictionary(ls_opinioni_tokenizzate)
corpus = [id2word.doc2bow(text) for text in ls_opinioni_tokenizzate]
lda_model = gensim.models.ldamodel.LdaModel.load(r'C:\Users\user\Desktop\UNIMI\ProjectInfoRet\data\lda_{}.model'.format(year))

pyLDAvis.enable_notebook()
LDAvis_prepared = pyLDAvis.gensim_models.prepare(lda_model, corpus, id2word)
LDAvis_prepared
```
